```python
#build the network from the facebook dataset
import networkx as nx
G = nx.read_edgelist("facebook_combined.txt", create_using = nx.Graph(), nodetype=int)
```

```python
print(nx.info(G))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

nx.draw(G, node_size=50)
plt.show()
```

```python
G_fb=G
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
```

```python
sorted(betCent, key=betCent.get, reverse=True)[:5]
```

```python
import collections
import matplotlib.pyplot as plt

degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # degree sequence
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt, width=0.80, color='b')

plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
ax.set_xticks([d + 0.4 for d in deg])
ax.set_xticklabels(deg)


plt.loglog(range(1,G.order()+1),degree_sequence,'k.')
plt.xlabel("Rank")
plt.ylabel("Degree")
plt.title("Loglog plot")
plt.show()
```

```python
r=nx.degree_assortativity_coefficient(G)
print("Degree assortativity:\n")
print(r)
```

```python
sp=nx.all_pairs_dijkstra_path(G)

pair=nx.all_pairs_shortest_path(G)
count=0
print("showing 20 shortest paths")
for n in pair:
    if(count<20):
        print(n,sep='\n')
    else:
        break            
    count+=1
```

```python
Cd=nx.degree_centrality(G)
Cc=nx.closeness_centrality(G)
Cb=nx.betweenness_centrality(G)
Ce=nx.eigenvector_centrality(G)
print("Degree Centrality:\n")
print(Cd)
print("\ncloseness Centrality:\n")
print(Cc)
print("\nBetweenness Centrality:\n")
print(Cb)
print("\nEigenvector Centrality:\n")
print(Ce)
```

```python

sorted_dc=sorted(Cd.items())
sorted_bc=sorted(Cb.items())
sorted_cc=sorted(Cc.items())
sorted_ec=sorted(Ce.items())

deg_data=[y for x,y in sorted_dc]
close_data=[y for x,y in sorted_cc]
betw_data=[y for x,y in sorted_bc]
eig_data=[y for x,y in sorted_ec]

from scipy.stats.stats import pearsonr
r1=pearsonr(deg_data,betw_data)
s= "\n Pearson Correlation Coeff deg and betweenness: " + repr(r1)
print (s)
r2=pearsonr(deg_data,close_data)
s= "\n Pearson Correlation Coeff deg and closeness: " + repr(r2)
print (s)
r3=pearsonr(betw_data,close_data)
s= "\n Pearson Correlation Coeff betweennness and closeness: " + repr(r3)
print (s)
r4=pearsonr(betw_data,eig_data)
s= "\n Pearson Correlation Coeff betweennness and eigenvector : " + repr(r4)
print (s)
```

```python

from operator import itemgetter
node_degree=G.degree()
(largesthub,degree)=sorted(node_degree,key=itemgetter(1))[-1]

three_largest_Cd = nlargest(3, Cd, key=Cd.get)
print(three_largest_Cd )

hub_ego1=nx.ego_graph(G,three_largest_Cd[0])
pos=nx.spring_layout(hub_ego1)
nx.draw(hub_ego1,pos,node_color='b',node_size=50,with_labels=True)
nx.draw_networkx_nodes(hub_ego1,pos,nodelist=[three_largest_Cd[0]],node_size=300,node_color='r')
plt.show()

hub_ego1=nx.ego_graph(G,three_largest_Cd[1])
pos=nx.spring_layout(hub_ego1)
nx.draw(hub_ego1,pos,node_color='b',node_size=50,with_labels=True)
nx.draw_networkx_nodes(hub_ego1,pos,nodelist=[three_largest_Cd[1]],node_size=300,node_color='r')
plt.show()

hub_ego1=nx.ego_graph(G,three_largest_Cd[2])
pos=nx.spring_layout(hub_ego1)
nx.draw(hub_ego1,pos,node_color='b',node_size=50,with_labels=True)
nx.draw_networkx_nodes(hub_ego1,pos,nodelist=[three_largest_Cd[2]],node_size=300,node_color='r')
plt.show()
```
